```python
import pandas as pd
import numpy as np
df=pd.read_csv("new_train.csv")
df.head()
```

```python
df.isnull().sum() # Checking if there is any null value in the dataset 
```

```python
#Creating a Label Encoder
from sklearn import preprocessing 
le = preprocessing.LabelEncoder() 
```

```python
# Performing Label Encoding on the necessary features
df['realtionship_status']= le.fit_transform(df['realtionship_status'])
df['industry']= le.fit_transform(df['industry'])
df['genre']= le.fit_transform(df['genre'])
df['targeted_sex']= le.fit_transform(df['targeted_sex'])
df['airtime']= le.fit_transform(df['airtime'])
df['airlocation']= le.fit_transform(df['airlocation'])
df['expensive']= le.fit_transform(df['expensive'])
df['money_back_guarantee']= le.fit_transform(df['money_back_guarantee'])
df['netgain']= le.fit_transform(df['netgain'])
```

```python
df.head()
```

```python
df2=df.drop(["id","genre","airlocation","expensive","money_back_guarantee","netgain"],axis=1)
```

```python
x=df2[:]
y=df["netgain"]
```

```python
# Performing Label Encoding on the necessary features
df1['realtionship_status']= le.fit_transform(df1['realtionship_status'])
df1['industry']= le.fit_transform(df1['industry'])
df1['genre']= le.fit_transform(df1['genre'])
df1['targeted_sex']= le.fit_transform(df1['targeted_sex'])
df1['airtime']= le.fit_transform(df1['airtime'])
df1['airlocation']= le.fit_transform(df1['airlocation'])
df1['expensive']= le.fit_transform(df1['expensive'])
df1['money_back_guarantee']= le.fit_transform(df1['money_back_guarantee'])
```

```python
df1.head()
```

```python
df3=df1.drop(["id","genre","airlocation","expensive","money_back_guarantee"],axis=1)
df3.head()
```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8)
```

```python
model = xgb.XGBClassifier(learning_rate =0.0052,
 n_estimators=1000,
 max_depth=4,
 min_child_weight=6,
 gamma=0,
 subsample=0.8,
 colsample_bytree=0.8,
 reg_alpha=0.005,
 objective= 'binary:logistic',
 nthread=4,
 scale_pos_weight=1,
 seed=27)
model.fit(x_train,y_train)
```

```python
y_test=model.predict(df3)
```

```python
submission_5 = pd.DataFrame({'id':df1['id'],'netgain':y_test})
#submission_5.head()
```
